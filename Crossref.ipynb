## Feeding in list of DOI and Obtaining the Types of License

```python
# Feed in a list of DOIs and return a table with the publisher, license, resource links information and save it into a CSV file (Desktop).
doi_list = ["10.1038/s41566-018-0310-5",
                "10.1364/OPTICA.5.001396",
                "10.1016/j.physa.2022.128435",
                "10.1142/S0217751X22500786",
                "10.3390/photonics10060700",
                "10.1038/s41586-022-05190-2"]

def get_field_data(doi_list):
    df_temp = pd.DataFrame(columns=['doi','publisher','license_datetime','license_content_version',
                                    'license_delayindays', 'license_URL', 'resource_URL', 'doc_URL'])
    for doi_num in doi_list:
        list_temp = []
        result = cr.works(ids = doi_num)
        try:
            #doi_lic_info = result['message'].get('license', [{}])[0]

            #doi_lic_dt = doi_lic_info.get('start', {}).get('date-time', "NA")
            doi_lic_dt = result['message']['license'][0]['start']['date-time']
            
            #doi_lic_contentver = doi_lic_info.get('content-version', "NA")
            doi_lic_contentver = result['message']['license'][0]['content-version']

            #doi_lic_delayindays = doi_lic_info.get('delay-in-days', "NA")
            doi_lic_delayindays = result['message']['license'][0]['delay-in-days']

            #doi_lic_URL = doi_lic_info.get('URL', "NA")
            doi_lic_URL = result['message']['license'][0]['URL']
            
            #doi_publisher = doi_lic_info.get('publisher', "NA")
            doi_publisher = result['message']['publisher']

            #doi_resource = doi_lic_info.get('resource', {}).get('primary', "NA")
            doi_resource = result['message']['resource']['primary']['URL']
            
            for i in range(0, len(result['message']['link'])):
                list_temp.append(result['message']['link'][i]['URL'])
            doi_docURL = list_temp

        except:
            doi_lic_dt = "NA"
            doi_lic_contentver = "NA"
            doi_lic_delayindays = "NA"
            doi_lic_URL = "NA"
            doi_publisher = result['message']['publisher']
            doi_resource = result['message']['resource']['primary']['URL']

            for i in range(0, len(result['message']['link'])):
                list_temp.append(result['message']['link'][i]['URL'])
            doi_docURL = list_temp

        doi_obtained = [doi_num,doi_publisher,doi_lic_dt, doi_lic_contentver, doi_lic_delayindays, doi_lic_URL, doi_resource, doi_docURL]
        df_temp.loc[len(df_temp)] = doi_obtained
        df_temp.to_csv('doi_output.csv', index=False)

    return df_temp

df_output = get_field_data(doi_list)
df_output
```
